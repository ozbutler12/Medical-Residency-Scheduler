# Creating the Dataframes

```python
def year1(data1):
  # Convert the data to a DataFrame
  df1 = pd.DataFrame(data1)

  # Set the index to be the month names
  df1.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

  # Transpose the DataFrame so that the names are in the first column
  df1 = df1.transpose()

  # Rename the index to be 'Name'
  df1.index.name = 'Name'

  num_to_rotation = {
      1: 'Inpt',
      2: 'OBGYN',
      3: 'Peds',
      4: 'NICU',
      5: 'Emergency Med',
      6: 'Psych',
      7: 'Surg'
  }
  # Replace numbers with letters using the map() method
  for ind in df1:
    df1[ind] = df1[ind].map(num_to_rotation)

  return df1
```

```python
def year2(data2):
  df2 = pd.DataFrame(data2)

  # Set the index to be the month names
  df2.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

  # Transpose the DataFrame so that the names are in the first column
  df2 = df2.transpose()

  # Rename the index to be 'Name'
  df2.index.name = 'Name'

  num_to_rotation = {
      1: 'Inpt',
      2: 'EM UMC',
      3: 'Gyn',
      4: 'Cards,',
      5: 'Crit Care',
      6: 'Amb Peds',
      7: 'Practice Mngmt',
      8: 'Peds EM',
      9: 'Ortho'
  }
  # Replace numbers with letters using the map() method
  for ind in df2:
    df2[ind] = df2[ind].map(num_to_rotation)

  # Print the DataFrame
  return df2
```

```python
def year3(data3):
  # Convert the data to a DataFrame
  df3 = pd.DataFrame(data3)

  # Set the index to be the month names
  df3.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

  # Transpose the DataFrame so that the names are in the first column
  df3 = df3.transpose()

  # Rename the index to be 'Name'
  df3.index.name = 'Name'

  num_to_rotation = {
      1: 'Elective',
      2: 'Inpt',
      3: 'ENT/Opthal',
      4: 'Uro/Neuro,',
      5: 'MSK',
      6: 'Amb Peds',
      7: 'Comm Med',
      8: 'Peds EM',
      9: 'Derm'
  }
  # Replace numbers with letters using the map() method
  for ind in df3:
    df3[ind] = df3[ind].map(num_to_rotation)

  # Print the DataFrame
  return df3
```
